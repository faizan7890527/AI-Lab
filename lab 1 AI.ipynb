```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
```

```python
i = 1
while i <= 10:
    print(i)
    i = i + 1

num = 1
while num <= 20:
    if num % 2 == 0:
        print(num)
    num = num + 1
```

```python

num = int(input("Enter a number: "))
digit_sum = sum(int(digit) for digit in str(num))
if num % digit_sum == 0:
    print(f"{num} is a Harshad Number.")
else:
    print(f"{num} is not a Harshad Number.")
```

```python
strings = ['abc', 'xyz', 'aba', '1221', 'xyzzyx']
i = 0

for s in strings:
    if len(s) >= 2 and s[0] == s[-1]:
        i += 1

print("Expected Result:", i)
```

```python

numbers = (1, 2, 3 ,4 ,5 ,6 )
sum_result = sum(numbers)
product = 1
for n in numbers:
    product *= n

print("Sum is:", sum_result)
print("Product is:", product)
```

```python
numbers = [1, 2, 3 , 4 , 5]

largest = numbers[0]
for n in numbers:
    if n > largest:
        largest = n

print("Largest number is:", largest)
```

```python

L = [(), (), ('',), ('a', 'b'), (), ('a', 'b', 'c'), (), ('d',)]
filtered_list = [t for t in L if t and 'a' not in t]
print("Original List:", L)
print("After Removing Empty Tuples and Tuples with 'a':", filtered_list)
```

```python

n = int(input("Enter a number: "))
result = {str(x): x * x for x in range(1, n + 1)}
print(result)
```

```python

num_words = {
    '0': 'Zero',
    '1': 'One',
    '2': 'Two',
    '3': 'Three',
    '4': 'Four',
    '5': 'Five',
    '6': 'Six',
    '7': 'Seven',
    '8': 'Eight',
    '9': 'Nine'
}


number = input("Enter a number: ")


result = ' '.join(num_words[digit] for digit in number if digit in num_words)


print("In words:", result)
```
