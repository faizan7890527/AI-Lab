```python

class Item:
    name = "Faizan"
    description = "Rich_Dad_Poor_Dad"
    price = 600

    def viewFullDescription(self):
        print("Name:", self.name)
        print("Description:", self.description)
        print("Price:", self.price)

    def addToShoppingBasket(self):
        print(self.name, "added to shopping basket.")

    def removeFromShoppingBasket(self):
        print(self.name, "removed from shopping basket.")



class MP3(Item):
    name = "Champions Anthem"
    description = "Punjabi Song"
    price = 120
    artist = "Karan Aujla"
    duration = "4 min"

    def play(self):
        print("Playing:", self.name)

    def download(self):
        print("Downloading:", self.name)


class DVD(Item):
    name = "The Gorge"
    description = "Action Movie"
    price = 500
    certificate = "PG-13"
    duration = "2 hours"
    actors = "Miles Teller,Anya Taylor-Joy"

    def viewTrailer(self):
        print("Viewing trailer of:", self.name)



class Book(Item):
    name = "Rich Dad Poor Dad"
    description = "Economics Book"
    price = 350
    author = " Robert Kiyosaki and Sharon Lechter"
    numberOfPages = 250
    genre = "Economic and business"

    def previewContent(self):
        print("Previewing book:", self.name)




mp3 = MP3()
dvd = DVD()
book = Book()

mp3.viewFullDescription()
mp3.play()
mp3.download()

dvd.viewFullDescription()
dvd.viewTrailer()

book.viewFullDescription()
book.previewContent()
```

```python
def circle():
    radius = float(input("Enter radius: "))
    perimeter = 2 * 3.1416 * radius
    print("Perimeter (Circumference) =", perimeter)


def area():
    length = float(input("Enter length: "))
    width = float(input("Enter width: "))
    area = length * width
    print("Area =", area)

area()
circle()
```

```python
def unique_list(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

# Example
sample_list = [1, 2, 3, 3, 4, 4, 5, 3, 7, 5]
print("Sample List:", sample_list)
print("Unique List:", unique_list(sample_list))

    
```

```python
print("Calculation of Factorial of a number")
m = int(input("Enter a number: "))
def factorial(n):
    if n < 0:
        print("invalid input")
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

result = factorial(m)
print(result)
```
